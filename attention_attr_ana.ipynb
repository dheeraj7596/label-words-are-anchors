```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

def plot_ndarray(data, out_pdf_name):
    names = [r'$S_{wp}$', r'$S_{pq}$', r'$S_{ww}$']
    type_num = len(data)
    
    
    fig, ax = plt.subplots()
    
    for i in range(type_num):
        ax.plot(data[i], label=names[i]) 
        
    ax.legend()
    ax.set_ylabel('S')
    ax.set_xlabel('Layer')
    
    fig.savefig(out_pdf_name)
    plt.show()
```

```python
in_file_path = "results/llama2-7b-gsm8k-firsttoken-question"
out_prop_pkl_file = "output_dir/sst2_prop.pkl"
out_pdf_name = "output_dir/llama2-7b-gsm8k-firsttoken-question"
normalize = False
```

```python
def load_result(in_file_path):
    with open(in_file_path, 'rb') as f:
        return pickle.load(f)
```

```python
proportions_list = []
for seed in [42]:
    results = load_result(in_file_path)
    proportions = results.mean(axis=0)
    if normalize:
        proportions = proportions / proportions.sum(axis=0,keepdims=True)
    proportions_list.append(proportions[np.newaxis,...])
proportions = np.concatenate(proportions_list,axis=0)
proportions = proportions.mean(0)

with open(out_prop_pkl_file, 'wb') as f:
    pickle.dump(proportions, f)
```

```python
plot_ndarray(proportions, out_pdf_name)
```
